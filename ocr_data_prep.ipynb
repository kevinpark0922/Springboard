```python
def find(name):
    try:
        f = open("data/ocr/" + name)
        return f
    except FileNotFoundError:
        print("")
```

```python
from re import match as re_match
def is_string_non_numeric(s):
    """ Returns False is string is a number. """
    if re_match("^\d+?\.\d+?$", s) is None:
        return not s.isdigit()
    return True
```

```python
import os
import json
from collections import Counter
wordfreq = {}
cnt = Counter()
f = open("data/balancesheet.txt","r")
contents = f.readlines()
#rint(contents)
for x in contents:
    ocr_file = find(x.rstrip())
    if ocr_file is not None:
        #print(type(ocr_file))
        data = json.load(ocr_file)
        for word in data:
            pure_word = word['ocr_text'].replace(',', '').replace('.', '').replace('$', '').replace('-', '').replace('(', '').replace(')', '')
            if pure_word is not '' and pure_word is not ' ':
                if is_string_non_numeric(pure_word):
                    cnt[pure_word] += 1
                    
        
    
```

```python
import os
import json
from collections import Counter
wordfreq = {}
bigram_cnt = Counter()
f = open("data/balancesheet.txt","r")
contents = f.readlines()
#rint(contents)
for x in contents:
    ocr_file = find(x.rstrip())
    if ocr_file is not None:
        #print(type(ocr_file))
        data = json.load(ocr_file)
        length = len(data)
        for i in range(0,length,1):
            if length > (i+1):
                w1 = data[i]['ocr_text'].replace(',', '').replace('.', '').replace('$', '').replace('%', '').replace('-', '').replace('(', '').replace(')', '')
                w2 = data[i+1]['ocr_text'].replace(',', '').replace('.', '').replace('$', '').replace('%', '').replace('-', '').replace('(', '').replace(')', '')
                #print("%s %s" % (w1, w2))
                if is_string_non_numeric(w1) and is_string_non_numeric(w2):
                    bigram = w1 +" " + w2
                    #print(bigram)
                    if bigram is not '' and bigram is not ' ' and bigram != '   ':
                        bigram_cnt[bigram] += 1
                
        
```

```python
import os
import json
from collections import Counter
wordfreq = {}
trigram_cnt = Counter()
f = open("data/balancesheet.txt","r")
contents = f.readlines()
#rint(contents)
for x in contents:
    ocr_file = find(x.rstrip())
    if ocr_file is not None:
        #print(type(ocr_file))
        data = json.load(ocr_file)
        length = len(data)
        for i in range(0,length,1):
            if length > (i+2):
                w1 = data[i]['ocr_text'].replace(',', '').replace('.', '').replace('%', '').replace('$', '').replace('-', '').replace('(', '').replace(')', '')
                w2 = data[i+1]['ocr_text'].replace(',', '').replace('.', '').replace('%', '').replace('$', '').replace('-', '').replace('(', '').replace(')', '')
                w3 = data[i+2]['ocr_text'].replace(',', '').replace('.', '').replace('%', '').replace('$', '').replace('-', '').replace('(', '').replace(')', '')
                #print("%s %s" % (w1, w2))
                if is_string_non_numeric(w1) and is_string_non_numeric(w2) and is_string_non_numeric(w3):
                    trigram = w1 + " " + w2 + " " + w3
                    #print(bigram)
                    if trigram is not '' and trigram is not ' ' and trigram != '   ' and trigram != '     ' and trigram != '  ':
                        trigram_cnt[trigram] += 1
```
